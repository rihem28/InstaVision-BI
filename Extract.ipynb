```python
import pandas as pd
import os

# ---- Define Staging Directory ----
staging_dir = 'data/staging'
os.makedirs(staging_dir, exist_ok=True)

# ---- Extracting Data ----
instagram_data = pd.read_csv(f'{staging_dir}/Instagram_Analytics.csv')

# ---- Save Raw Data to Staging Area ----
instagram_data.to_csv(f'{staging_dir}/instagram_raw.csv', index=False)

print("Instagram data extraction completed and saved to staging area.")

```
